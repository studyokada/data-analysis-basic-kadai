```python
import matplotlib.pyplot as plt
import numpy as np


fig, ax = plt.subplots(2, 3,tight_layout=True)



x = np.linspace(-10, 10, 100)

y = [1, 2, -5, 2]
ax[0, 0].plot(y)

y = np.sin(x)
ax[0, 1].plot(x, y)


y = x**2
ax[1, 1].plot(x, y)

y = 2 * x
ax[1, 2].plot(x, y)

x = np.random.normal(15, 5, 2000)
ax[0, 2].hist(x)

np.random.seed(3)
x = np.random.normal(0, 2, 60)
y = np.random.normal(0, 2, len(x))
ax[1, 0].scatter(x, y)

plt.show()

```
